```python
from bs4 import BeautifulSoup
import requests

head_Html_lotto='https://web.cpc.com.tw/division/mb/oil-more4.aspx'
res = requests.get(head_Html_lotto, timeout=30)
print(res.text)
```

```python
import pandas as pd
soup = BeautifulSoup(res.text)
a = soup.findAll(id="Showtd")
b=a[0].findAll("tr")
alldata=[]
for i in b:
    c=i.findAll("td")
    data=[]
    for j in c:   
        data.append(j.text.strip())
    data=data[:5]
    alldata.append(data)
column_name=["date","無鉛汽油92","無鉛汽油95","無鉛汽油98","超級/高級柴油"]
df=pd.DataFrame(columns=column_name)
data=[]
for i in alldata:
    data.append({"date":i[0],"無鉛汽油92":i[1],"無鉛汽油95":i[2],"無鉛汽油98":i[3],"超級/高級柴油":i[4]})

data.reverse() 
data=data[8:]
df=pd.concat([pd.DataFrame(data), df], ignore_index=True,sort=True)
df["date"] = pd.to_datetime(df["date"])
print(df)
```

```python
df=df.convert_objects(convert_numeric=True)
df=df.fillna(0)
print(df)
```

```python
import pandas as pd
place=["雲林","嘉義","彰化","台南","高雄","屏東","台中","苗栗","桃園","台北"]
column_list=["無鉛汽油92","無鉛汽油95","無鉛汽油98","超級/高級柴油"]
crop_list=["high",'low','mean','medium','volume','date']
place_name=["train+oil+全weather+國定假日+拜拜日.csv"]
for i in place_name:
    a=pd.read_csv("文旦trend.csv")
    #b=pd.read_csv("甘藍trend.csv")
    train=pd.read_csv(i)
    a=a.drop(["Unnamed: 0"], axis=1)
    a=a.drop(["index"], axis=1)
    #b=b.drop(["Unnamed: 0"], axis=1)
    train=train.drop(["Unnamed: 0"], axis=1)
    train=train.merge(a, on='date', how='left')
    train=train.fillna(0)
    train=train.drop(["high"], axis=1)
    train=train.drop(["low"], axis=1)
    train=train.drop(["mean"], axis=1)
    train=train.drop(["medium"], axis=1)
    train=train.drop(["volume"], axis=1)
    b=pd.read_csv("柚子-文旦.csv")
    b=b.drop(["Unnamed: 0"], axis=1)
    b=b[crop_list]
    train=train.merge(b, on='date', how='left')
    train=train.fillna(0)
    train.to_csv("train+oil+全weather+國定假日+拜拜日+文旦trend.csv", encoding='utf_8_sig')
```
